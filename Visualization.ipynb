```python
#imported all necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb
import math
```

```python
#importing csv file
df=pd.read_csv('epi_r.csv')
df
```

```python
df.info()
```

```python
#describing statistical information about the dataset like mean,meax and min value
df.describe()
```

```python
df.nunique()
```

```python
#count of all null values in the calories sum
df.calories.isnull().sum()
```

```python
#eliminate the rows where calories are more than 10000
df=df[(df['calories']<10000) | (df['calories'].isnull==1)]
```

```python
#mean of calories column
calorie_mean=df.calories.mean()
calorie_mean
```

```python
#filling the empty values with mean value
df.calories.fillna(calorie_mean,inplace=True)
```

```python
df.calories.describe()
```

```python
df.rating.isnull().sum()
```

DATA VISUALIZATION

different visualization for rating column

```python
sb.scatterplot(data=df['rating'])
```

```python
sb.pairplot(data=df.iloc[:,1:3])
```

```python
sb.kdeplot(df["rating"], color = "red",
            shade = True)
plt.show()
plt.figure()
```

```python
#histogram to observe the frequency of different ratings
df.rating.hist(bins=30)
plt.title('Recipe Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python

# Box plot to visualize the distribution of ratings
plt.figure(figsize=(8, 6))
sb.boxplot(x='rating', data=df)
plt.title('Box Plot of Recipe Ratings')
plt.show()
```

```python
# Violin plot to visualize the distribution of ratings with more detail
plt.figure(figsize=(8, 6))
sb.violinplot(x='rating', data=df)
plt.title('Violin Plot of Recipe Ratings')
plt.show()
```

```python
# Scatter plot to see the relationship between calories and rating
plt.figure(figsize=(8, 6))
plt.scatter(df['calories'], df['rating'])
plt.xlabel('Calories')
plt.ylabel('Rating')
plt.title('Calories vs Rating')
plt.show()

```

```python
# Histogram of calories
plt.figure(figsize=(8, 6))
df.calories.hist(bins=30)
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()

```

```python

# Box plot of calories
plt.figure(figsize=(8, 6))
sb.boxplot(x='calories', data=df)
plt.title('Box Plot of Calories')
plt.show()
```

```python
# Violin plot of calories
plt.figure(figsize=(8, 6))
sb.violinplot(x='calories', data=df)
plt.title('Violin Plot of Calories')
plt.show()
```
